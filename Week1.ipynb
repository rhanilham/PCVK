````python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # For image display
from skimage import io, transform
from PIL import Image
import matplotlib.pylab as plt
````

````python
# Membuat list untuk menyimpan URL dari beberapa image
urls = ["https://1.bp.blogspot.com/-Dtb8C8vyjVk/UpFcVls1I_I/AAAAAAAAAHU/PgyXdNhNyxQ/s1600/Desain-Sekolah-Modern-3-Lantai.jpg",
    "https://i.pinimg.com/736x/25/19/08/2519080941a30f29ddef6413ff447d4d--benny-the-bull-chicago-bulls.jpg",
    "https://cdn-cas.orami.co.id/parenting/images/pexels-mati-mango-4734723.width-800.jpg"]
# Baca dan tampilkan image
# Loop pada tiap URL image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)  # Baca image
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)  # Ubah ukuran image menjadi setengah ukuran
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # Konversi warna ke RGB
    final_frame = cv.hconcat((image, image_2))  # Gabungkan image
    cv2_imshow(final_frame)  # Tampilkan image
    print('\n')
````

````python
# Mendapatkan tinggi dan lebar dari image_2
tinggi = image_2.shape[0]
lebar = image_2.shape[1]

# Mencetak resolusi gambar
print("Resolusi image: {} x {}".format(tinggi, lebar))

# Menampilkan gambar
cv2_imshow(image_2)
````

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
# Membuat garis horizontal di tengah image
for y in range(lebar):
    image_3[int(tinggi / 2), y] = [255, 255, 255]
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
# Membuat garis menyilang diagonal dari kiri atas ke kanan bawah
for i in range(min(320, 400)):
    image_3[i, int(i * 400 / 320)] = [255, 255, 255]
# Membuat garis menyilang diagonal dari kiri bawah ke kanan atas
for i in range(min(320, 400)):
    image_3[320 - i - 1, int(i * 400 / 320)] = [255, 255, 255]
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
# Panjang garis horizontal yang diinginkan
panjang_garis = 150
# Hitung titik awal dan akhir garis horizontal
tengah_y = int(tinggi / 2)
tengah_x_start = max(0, int((lebar - panjang_garis) / 2))
tengah_x_end = min(lebar, tengah_x_start + panjang_garis)
# Buat garis horizontal di tengah gambar
image_3[tengah_y, tengah_x_start:tengah_x_end] = [255, 255, 255]
# Gabungkan gambar asli dan gambar dengan garis
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

````python
import requests
from io import BytesIO

# URL gambar
url_ktp = "https://github.com/user-attachments/assets/d21415fe-da79-43cf-a64a-8f120a26b24d"

# Mengunduh gambar dari URL
response = requests.get(url_ktp)
response.raise_for_status()  # Memastikan permintaan berhasil

# Membaca gambar dari respons
image = Image.open(BytesIO(response.content))

# Menyimpan gambar ke local
local_path = "gambar_ktp.jpg"
image.save(local_path)

# Membaca gambar menggunakan OpenCV
gambar_ktp = cv.imread(local_path)

# Konversi gambar dari BGR (format default OpenCV) ke RGB
gambar_ktp = cv.cvtColor(gambar_ktp, cv.COLOR_BGR2RGB)

# Fungsi untuk menggambar kotak
def draw_rectangle(image, pos_x, pos_y, width, height, color):
    """
    Menambahkan kotak ke gambar dengan posisi, ukuran, dan warna yang ditentukan.

    Parameters:
        image (numpy.ndarray): Gambar yang akan dimodifikasi.
        pos_x (int): Posisi horizontal titik kiri atas kotak.
        pos_y (int): Posisi vertikal titik kiri atas kotak.
        width (int): Lebar kotak.
        height (int): Tinggi kotak.
        color (tuple): Warna kotak dalam format BGR (Blue, Green, Red).
    """
    # Mendapatkan ukuran gambar
    tinggi, lebar = image.shape[:2]

    # Menghitung posisi akhir kotak
    start_x = max(pos_x, 0)
    end_x = min(pos_x + width, lebar)
    start_y = max(pos_y, 0)
    end_y = min(pos_y + height, tinggi)

    # Menambahkan kotak ke gambar
    image[start_y:end_y, start_x:end_x] = color

# Menutup RT/RW dengan kotak merah
draw_rectangle(gambar_ktp, pos_x=230, pos_y=180, width=70, height=15, color=[255, 0, 0])

# Menutup Kel/Desa dengan kotak hijau
draw_rectangle(gambar_ktp, pos_x=230, pos_y=198, width=100, height=15, color=[0, 255, 0])

# Menampilkan gambar menggunakan Matplotlib
plt.imshow(gambar_ktp)
plt.axis('off')
plt.title('Gambar KTP')
plt.show()
````
